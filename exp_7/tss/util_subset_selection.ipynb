```python
import pickle
from submodlib import FacilityLocationFunction
from submodlib import GraphCutFunction
from submodlib import DisparitySumFunction
from submodlib import DisparityMinFunction
from submodlib import LogDeterminantFunction
```

```python
with open('ft_op_embeddings.pkl', 'rb') as f:
    finetune_data_op_embeddings = pickle.load(f)
```

```python
index_list_facloc={}
for i in range(0,len(finetune_data_op_embeddings), 100):
    print(int(i/100))
    indices=[]
    features = finetune_data_op_embeddings[i:i+100]
    objFL = FacilityLocationFunction(n=100, data=features, mode="dense", metric="cosine")
    greedyList = objFL.maximize(budget=3,optimizer='NaiveGreedy', stopIfZeroGain=False, stopIfNegativeGain=False, verbose=False)
    indices.append([i[0] for i in greedyList])
    index_list_facloc[int(i/100)] = indices
```

```python
index_list_gc={}
for i in range(0,len(finetune_data_op_embeddings), 100):
    print(int(i/100))
    indices=[]
    features = finetune_data_op_embeddings[i:i+100]
    objFL = GraphCutFunction(n=100, mode="dense", lambdaVal=0.5, data=features, metric="cosine")
    greedyList = objFL.maximize(budget=3,optimizer='NaiveGreedy', stopIfZeroGain=False, stopIfNegativeGain=False, verbose=False)
    print(greedyList)
    indices.append([i[0] for i in greedyList])
    index_list_gc[int(i/100)] = indices
```

```python
index_list_disparitysum={}
for i in range(0,len(finetune_data_op_embeddings), 100):
    print(int(i/100))
    indices=[]
    features = finetune_data_op_embeddings[i:i+100]
    objFL = DisparitySumFunction(n=100, data=features, mode="dense", metric="cosine")
    greedyList = objFL.maximize(budget=3,optimizer='NaiveGreedy', stopIfZeroGain=False, stopIfNegativeGain=False, verbose=False)
    #print(greedyList)
    indices.append([i[0] for i in greedyList])
    index_list_disparitysum[int(i/100)] = indices
```

```python
index_list_disparitymin={}
for i in range(0,len(finetune_data_op_embeddings), 100):
    print(int(i/100))
    indices=[]
    features = finetune_data_op_embeddings[i:i+100]
    objFL = DisparityMinFunction(n=100, data=features, mode="dense", metric="cosine")
    greedyList = objFL.maximize(budget=3,optimizer='NaiveGreedy', stopIfZeroGain=False, stopIfNegativeGain=False, verbose=False)
    #print(greedyList)
    indices.append([i[0] for i in greedyList])
    index_list_disparitymin[int(i/100)] = indices
```

```python
index_list_logdeterminant={}
for i in range(0,len(finetune_data_op_embeddings), 100):
    print(int(i/100))
    indices=[]
    features = finetune_data_op_embeddings[i:i+100]
    objFL = LogDeterminantFunction(n=100, data=features, mode="dense", metric="cosine", lambdaVal=0.5)
    greedyList = objFL.maximize(budget=3,optimizer='NaiveGreedy', stopIfZeroGain=False, stopIfNegativeGain=False, verbose=False)
    #print(greedyList)
    indices.append([i[0] for i in greedyList])
    index_list_logdeterminant[int(i/100)] = indices
```
